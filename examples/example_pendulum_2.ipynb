```python
import numpy as np
import gym
import matplotlib.pyplot as plt
from IPython.display import clear_output
from tqdm import trange

import actor_critic
import distributions
```

```python
environ_name = "Pendulum-v0"
env = actor_critic.EnvBatch( lambda : actor_critic.make_env(environ_name, max_steps=500 ), n_envs=10 )
```

```python
actor = actor_critic.ActorMLP(env=env, scale_model_inputs=False, activation='elu')
critic = actor_critic.CriticMLP(env=env, scale_model_inputs=False, activation='elu')

ac = actor_critic.ActorCritic(
                                 env=env,
                                 actor=actor, 
                                 critic=critic,
                                 entropy_coef=0.1,
                                 actor_learning_rate=0.0001,
                                 critic_learning_rate=0.001,
                                 gamma=0.98
                             )
```

```python
ac.train_and_display(n_iters=10001, n_games_per_plot_update=5, stop_score=500 )
```
